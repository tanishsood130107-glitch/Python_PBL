```python
df["SPENDING"] = df["QUANTITY"]*df["UNIT_PRICE"]
```

```python
df["Returned_Amount"] = df["UNIT_PRICE"]*df["NO_OF_ITEMS_RETURNED"]
```

```python
df.head()
```

```python
df["CUSTOMER_ID"].nunique()
```

```python
df_1 = df.groupby("CUSTOMER_ID")["SPENDING"].sum()
```

```python
df1 = pd.DataFrame(df_1)
```

```python
df1.head()
```

```python
df_2 = df.groupby("CUSTOMER_ID")["INVOICE_NO"].nunique()
```

```python
df["TotalPrice"] = df["QUANTITY"] * df["UNIT_PRICE"]
```

```python
df.head()
```

```python
df_3 = df.groupby("CUSTOMER_ID").agg({
    'SPENDING': 'sum',
    'INVOICE_NO': 'nunique',
    'QUANTITY': 'sum'
}).rename(columns={
    'TOTAL_PRICE': 'Total_Spending',
    'INVOICE_NO': 'FREQUENCY',
    'QUANTITY': 'TOTAL_QUANTITY'
}).reset_index()
```

```python
df_3.head()
```

```python
df_4 = df.groupby('CUSTOMER_ID').agg({
    'QUANTITY': 'sum',
    'NO_OF_ITEMS_RETURNED': 'sum',
}).reset_index()

```

```python
df_4.head()
```

```python
df_4['Return_Ratio'] = (
    df_4['NO_OF_ITEMS_RETURNED'] /
    df_4['QUANTITY']
)


```

```python
df_4.head()
```

```python
df_5 = df.groupby('CUSTOMER_ID')['PRODUCT_CATEGORY'].apply(list).reset_index()

df_5.head()
```

```python
df_5 = df.groupby('CUSTOMER_ID')['PRODUCT_CATEGORY'] \
               .apply(lambda x: sorted(set(x))) \
               .reset_index()

df_5.head()
```

```python
df_top2 = df.groupby('CUSTOMER_ID')['PRODUCT_CATEGORY'] \
    .agg(lambda x: x.value_counts().idxmax()) \
    .reset_index()

df_top2.head()
```

```python
df_top2["PRODUCT_CATEGORY"].value_counts()
```

```python
merged_df = pd.merge(df_3, df_4, on='CUSTOMER_ID', how='inner')
merged_df.head()
```

```python
df_raw = pd.read_csv("FINAL_PBL_DATA_SET_7.csv")   

df_raw['INVOICE_DATE'] = pd.to_datetime(df_raw['INVOICE_DATE'])

last_purchase = df_raw.groupby('CUSTOMER_ID')['INVOICE_DATE'].max().reset_index()
last_purchase = last_purchase.rename(columns={'INVOICE_DATE': 'Last_Purchase_Date'})
reference_date = df_raw['INVOICE_DATE'].max() + pd.Timedelta(days=1)

last_purchase['Recency'] = (reference_date - last_purchase['Last_Purchase_Date']).dt.days



```

```python
merged_df1 = pd.merge(merged_df,last_purchase,on='CUSTOMER_ID', how='inner')
merged_df1.head()
```

```python
final_df1 = pd.merge(merged_df1,df_top2,on='CUSTOMER_ID', how='inner')
final_df1.head()
```

```python
final_df = pd.merge(final_df1,avg_pur_gap, on = "CUSTOMER_ID" , how = "inner")
final_df.head()
```

```python
final_df.drop(columns = "Last_Purchase_Date" , inplace = True)
```
