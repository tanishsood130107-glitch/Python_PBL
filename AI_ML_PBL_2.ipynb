```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("FINAL_PBL_DATA_SET_7.csv" )
```

```python
df.head()
```

```python
df["SPENDING"] = df["QUANTITY"]*df["UNIT_PRICE"]
```

```python
df["Returned_Amount"] = df["UNIT_PRICE"]*df["NO_OF_ITEMS_RETURNED"]
```

```python
df.head()
```

```python
df["CUSTOMER_ID"].nunique()
```

```python
df_1 = df.groupby("CUSTOMER_ID")["SPENDING"].sum()
```

```python
df1 = pd.DataFrame(df_1)
```

```python
df1.head()
```

```python
df_2 = df.groupby("CUSTOMER_ID")["INVOICE_NO"].nunique()
```

```python
df["TotalPrice"] = df["QUANTITY"] * df["UNIT_PRICE"]
```

```python
df.head()
```

```python
df_3 = df.groupby("CUSTOMER_ID").agg({
    'SPENDING': 'sum',
    'INVOICE_NO': 'nunique',
    'QUANTITY': 'sum'
}).rename(columns={
    'TOTAL_PRICE': 'Total_Spending',
    'INVOICE_NO': 'FREQUENCY',
    'QUANTITY': 'TOTAL_QUANTITY'
}).reset_index()
```

```python
df_3.head()
```

```python
df_4 = df.groupby('CUSTOMER_ID').agg({
    'QUANTITY': 'sum',
    'SPENDING': 'sum',
    'NO_OF_ITEMS_RETURNED': 'sum',
    "Returned_Amount" : 'sum'
}).reset_index()

```

```python
df_4.head()
```

```python
df_4['Return_Ratio'] = (
    df_4['NO_OF_ITEMS_RETURNED'] /
    df_4['QUANTITY']
)


```

```python
df_4['Return_Amaount_Ratio'] = (
    df_4['SPENDING'] /
    df_4['Returned_Amount']
)
```

```python
df_4["Return_Amaount_Ratio"] = df_4["Return_Amaount_Ratio"].replace(float('inf'), 0)
```

```python
df_4.head()
```

```python
df_5 = df.groupby('CUSTOMER_ID')['PRODUCT_CATEGORY'].apply(list).reset_index()

df_5.head()
```

```python
df_5 = df.groupby('CUSTOMER_ID')['PRODUCT_CATEGORY'] \
               .apply(lambda x: sorted(set(x))) \
               .reset_index()

df_5.head()
```

```python
df_top2 = df.groupby('CUSTOMER_ID')['PRODUCT_CATEGORY'] \
    .apply(lambda x: x.value_counts().head(2).index.tolist()) \
    .reset_index()

df_top2.head()
```

```python
merged_df = pd.merge(df_3, df_4, on='CUSTOMER_ID', how='inner')
merged_df.head()
```

```python
final_df = pd.merge(merged_df,df_top2,on='CUSTOMER_ID', how='inner')
final_df.head()
```

```python
type(final_df)
```

```python
final_df = final_df.drop('SPENDING_y', axis=1)
final_df = final_df.drop('QUANTITY', axis=1)
```

```python
final_df.head()
```

```python
final_df = final_df.rename(columns={'SPENDING_x': 'TOTAL_SPENDING', "Return_Amaount_Ratio": "RETURN_AMOUNT_RATIO" ,  "Return_Ratio":"RETURN_RATIO","Returned_Amount":"RETURNED_AMOUNT", "PRODUCT_CATEGORY" : "TOP_CATEGORY"})
```
